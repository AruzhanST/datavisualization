```python
import os
import json
import re
from collections import defaultdict
import collections
DATA = ".\\abst"
punkt = re.compile(r"[:.,;!?-]+")
```

```python
paths = None
papers = []
for root,dirs,files in os.walk(".\\abst"):
    paths = [os.path.join(DATA, flname) for flname in files]
    
for path in paths:
    with open(path,"r", encoding="utf-8") as fh:
        papers.extend(json.load(fh))
print(len(papers));
```

```python
abstracts = [paper["abstract"] for paper in papers]
lengs = list(set([len(abs1) for abs1 in abstracts]))
sorted(lengs)
```

```python
abstracts_tok = [re.sub(punkt,"",abs1.replace("\n", " ")).lower().split(" ") for abs1 in abstracts]
lengs_abstracts_tok = [len(abs1) for abs1 in abstracts_tok]

print(sorted(lengs_abstracts_tok))
```
